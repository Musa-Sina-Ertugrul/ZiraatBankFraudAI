```python
from translate import Translator
translator= Translator(to_lang="Turkish")
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/musasina/Desktop/projects/ZiraatBankFraudAI/datasets/job_descriptions.csv")["Job Title"].convert_dtypes(convert_string=True)
```

```python
df = df.dropna()
```

```python
jobs = []

for job in set(df.to_list()):
    print(job)
    jobs.append(translator.translate(job))

jobs
```

```python
pd.DataFrame(data=jobs,columns=["names"]).to_csv("/home/musasina/Desktop/projects/ZiraatBankFraudAI/datasets/job_list.csv")
```
